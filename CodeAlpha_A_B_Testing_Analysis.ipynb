```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as ss
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("data.csv")

# Display the first few rows of the dataframe
print(data.head())

# Plot the distribution of the target and control groups
sns.distplot(data.Conversion_A)
sns.distplot(data.Conversion_B)
plt.legend(['Conversion_A', 'Conversion_B'])
plt.show()

# Perform t-test
t_stat, p_val = ss.ttest_ind(data.Conversion_B, data.Conversion_A)

# Print the t-statistic and p-value
print("t-statistic:", t_stat)
print("p-value:", p_val)

# Determine significance
alpha = 0.05
if p_val < alpha:
    print("Reject null hypothesis: There is a statistically significant difference between the two groups.")
else:
    print("Fail to reject null hypothesis: There is no statistically significant difference between the two groups.")
```
